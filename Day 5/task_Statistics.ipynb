```python
# Statistics
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#a
fig=plt.figure(figsize=(20,10))
plt.subplot(1,3,1)
x = np.arange(10)
mu = 2
r_pmf = stats.poisson.pmf(x, mu)
plt.stem(x,r_pmf)
plt.title('probability mass function')
plt.subplot(1,3,2)
r_cdf = stats.poisson.cdf(x, mu)
plt.stem(x,r_cdf)
plt.title('cummulative distribution function')
plt.subplot(1,3,3)
rvs = stats.poisson.rvs(mu, size=1000)
plt.hist(rvs, x)
plt.title('histogram of 1000 random realizations of the variable')
plt.show()
```

```python
# b
fig=plt.figure(figsize=(20,10))
plt.subplot(1,3,1)
x = np.linspace(-5,5,1000)
mu = 2
sigma = 1
r_pmf = stats.norm.pdf(x, mu, sigma)
plt.plot(x,r_pmf)
plt.title('probability mass function')
plt.subplot(1,3,2)
r_cdf = stats.norm.cdf(x, mu, sigma)
plt.plot(x,r_cdf)
plt.subplot(1,3,3)
plt.title('cummulative distribution function')
rvs = stats.norm.rvs(mu, sigma, size=1000)
plt.hist(rvs,x)
plt.title('histogram of 1000 random realizations of the variable')
plt.show()
```

```python
#c
test_res = stats.ttest_ind(stats.poisson.rvs(mu, size=100000),stats.poisson.rvs(mu, size=100000))
print(test_res)
```
